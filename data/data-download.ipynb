```python
import yfinance as yf
import pandas as pd
```

```python
TICKERS = {"CORN": "corn", 
           "FCG": "gas",
           "ZGLDUS.SW": "gold",
           "BTC-USD": "bitcoin"
           }
```

```python
data = pd.concat(
    [
        (yf.Ticker(ticker).history(period="max")
         .assign(date=lambda df: df.index.date)
         .set_index("date")
         ["Close"]
         .rename(ticker)) 
        for ticker in TICKERS.keys()
        ],
    axis=1
)
data
```

```python
(data
 .rename(columns=TICKERS)
 .dropna()
 .to_pickle("data.pkl")
 )
```
